```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("d.csv")
df
```

```python
%pylab inline
from pylab import *

for idx in range(9):
    print df.iloc[[idx]]["Taxes Owed"][idx].replace
    taxes_owed = float(df.iloc[[idx]]["Taxes Owed"][idx].replace(",", ""))
    market_cap = float(df.iloc[[idx]]["Market Cap"][idx])
    name = df.iloc[[idx]]["Name"][idx]

    figure(1, figsize=(6,6))
    ax = axes([0.1, 0.1, 0.8, 0.8])
    labels = 'Taxes Owed', ''
    fracs = [taxes_owed, market_cap - taxes_owed]
    explode=(0, 0.05)
    pie(fracs, explode=explode, labels=labels,
                autopct='%1.1f%%', shadow=False, startangle=90,
                colors=["#E4AB8E", "#B8C6D3"]
       )


    title(name)

    show()
```
